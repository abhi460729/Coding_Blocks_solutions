```python
titanic = pd.read_csv("Train.csv")
```

```python
round(titanic.describe(include="all" , percentiles =[0.1,0.25,0.75,0.99]))
```

```python
titanic = titanic.drop(["name" ,"pclass", "ticket" , "embarked" ,"cabin" ,'boat' ,"home.dest","body" ,"parch","sibsp"],axis=1)
```

```python
titanic
```

```python
titanic.info()
```

```python
titanic.age.isnull()
```

```python
titanic['age'] = titanic['age'].fillna(titanic['age'].mean())
```

```python
titanic.info()
```

```python
titanic.describe()
```

```python
titanic['sex'].replace(['male','female'],[0,1] , inplace=True)
```

```python
titanic
```

```python
titanic.isna().sum()
```

```python
titanic['fare'] = titanic['fare'].fillna(titanic['fare'].mean())
```

```python
titanic.isna().sum()
```

```python
titanic
```

```python
from sklearn.model_selection import train_test_split

train,test = train_test_split(titanic , test_size=0.3 ,random_state=0,stratify=titanic['survived'] )
```

```python
print(train.shape)
print(test.shape)
```

```python
train_X = train[train.columns[1:]]
train_Y = train[train.columns[:1]]
test_X = test[test.columns[1:]]
test_Y = test[test.columns[:1]]
```

```python
train_X
```

```python
train_Y.head()
```

```python
test_X
```

```python
from sklearn.linear_model import LogisticRegression
model = LogisticRegression()
```

```python
model.fit(train_X , train_Y)
```

```python
prediction = model.predict(test_X)
```

```python
prediction
```

```python
import pandas as pd 
pd.DataFrame(prediction).to_csv("file.csv")
```
