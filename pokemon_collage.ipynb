```python
red = np.zeros((50,50,3),dtype='int')
red[:,:,0] = 200
plt.imshow(red)
plt.show()

green = np.zeros((50,50,3),dtype='int')
green[:,:,1] = 200
plt.imshow(green)
plt.show()
```

```python
red_green = np.hstack((red,green))
green_red = red_green[:,::-1,:]


img = np.vstack((red_green,green_red))
plt.imshow(img)
plt.show()
```
